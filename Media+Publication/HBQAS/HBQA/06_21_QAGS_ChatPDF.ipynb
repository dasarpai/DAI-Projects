# Ask Questions - Interact with PdfGPT

```python
def pdfgpt_by_section(input, srcid, section_id):
  import requests

  headers = {
      'x-api-key': CHATGPT_KEY,
      "Content-Type": "application/json",
  }

  data = {
      "referenceSources": True,
      'sourceId': srcid,
      'messages': [
          {
              'role': "assistant",
              'content': f"go to section {section_id}",
              'role': "user",
              'content': input,
          }
      ]
  }

  response = requests.post(
      'https://api.chatpdf.com/v1/chats/message', headers=headers, json=data)

  if response.status_code == 200:
      print('Result:', response.json()['content'])
      return response.json()['content']
  else:
      print('Status:', response.status_code)
      print('Error:', response.text)
      return None
```

```python
section_id="L"
pdfgpt_by_section(f"Generate 3 questions, their corresponding answer and reference from section {section_id}?",srcid, section_id)
```

```python
section_id="V"
pdfgpt_by_section(f"Generate 3 questions, their corresponding answer and reference from section {section_id}?",srcid, section_id)
```

```python
def pdfgpt_by_pageno(input, srcid, page_number):
  import requests

  headers = {
      'x-api-key': CHATGPT_KEY,
      "Content-Type": "application/json",
  }

  data = {
      "referenceSources": True,
      'sourceId': srcid,
      'messages': [
          {
              'role': "system",
              'content': f"page number information is on the top of every page, go to page {page_number}",
              'role': "user",
              'content': input,
              "page_number": page_number,
          }
      ]
  }

  response = requests.post(
      'https://api.chatpdf.com/v1/chats/message', headers=headers, json=data)

  if response.status_code == 200:
      print('Result:', response.json()['content'])
      return response.json()['content']
  else:
      print('Status:', response.status_code)
      print('Error:', response.text)
      return None
```
