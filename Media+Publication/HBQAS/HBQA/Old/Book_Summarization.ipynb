# Roman to Integer


```python
# This function returns value of
# each Roman symbol
def value(r):
    if (r == 'I'):
        return 1
    if (r == 'V'):
        return 5
    if (r == 'X'):
        return 10
    if (r == 'L'):
        return 50
    if (r == 'C'):
        return 100
    if (r == 'D'):
        return 500
    if (r == 'M'):
        return 1000
    return -1

def romanToDecimal(txt):
    txt = txt.replace('.','')
    res = 0
    i = 0

    while (i < len(txt)):

        # Getting value of symbol s[i]
        s1 = value(txt[i])

        if (i + 1 < len(txt)):

            # Getting value of symbol s[i + 1]
            s2 = value(txt[i + 1])

            # Comparing both values
            if (s1 >= s2):

                # Value of current symbol is greater
                # or equal to the next symbol
                res = res + s1
                i = i + 1
            else:

                # Value of current symbol is greater
                # or equal to the next symbol
                res = res + s2 - s1
                i = i + 2
        else:
            res = res + s1
            i = i + 1

    return f"{res:03d}"

# Driver code
print("Integer form of Roman Numeral is"),
print(romanToDecimal("MCMIV"))
```

# Create Mahabharat Metadata File  

```python
import os 
SectionName=[]
SectionWordCount=[]
SectionParaCount=[]
SectionCharCount=[]
SectionType=[]

for file in os.listdir(r"MBBook-Sections//"):
    filename = r"MBBook-Sections//" + file
    with open(filename, "r") as txtfile:
        content = txtfile.readlines()
        paracount = len(content)
        charcount = len(" ".join(content))
        wordcount = len(" ".join(content).split(" "))
        
        SectionName.append(file)
        SectionWordCount.append(wordcount)
        SectionParaCount.append(paracount)
        SectionCharCount.append(charcount)

        if "MBBook" in file:
            SectionType.append('Book')
        else:
            SectionType.append('Section')

        #print(SectionName, SectionCharCount, SectionParaCount, SectionWordCount, SectionType)
        txtfile.close()


content = ""
paracount = ""
charcount = ""
wordcount = ""
           
```
