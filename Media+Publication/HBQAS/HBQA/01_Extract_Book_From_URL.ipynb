# Get Metadata of Files

```python
df['Section_Word_Count']=0
df['Section_Letter_Count']=0
df['Section_Para_Count']=0

# Open the URL
for row in range(df.shape[0]):
    filename = conf['CORPUS_SECTIONS_FOLDER'] + df.loc[row,'File_Name']
    with open(filename, "r") as file1:
        content = file1.readlines()
        para_count = len(content)
        content = "\n".join(content)
        letter_count = len(content)
        word_count = len(content.split(" "))
        df.loc[row,"Section_Word_Count"] = word_count
        df.loc[row,"Section_Letter_Count"] = letter_count
        df.loc[row,"Section_Para_Count"] = para_count
        print("Updating file: "+filename, letter_count, word_count, para_count )
```
