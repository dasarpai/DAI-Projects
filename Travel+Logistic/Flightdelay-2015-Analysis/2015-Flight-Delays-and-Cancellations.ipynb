```python
import pandas as pd
import numpy as np

def impute_nulls(df):
    """Imputes null values based on column data types."""
    for col in df.columns:
        if df[col].isnull().any():
            if pd.api.types.is_numeric_dtype(df[col]):
                if (df[col].isnull().sum() / len(df)) < 0.20:  # Less than 20% nulls
                    df[col] = df[col].fillna(df[col].median())
                # else: # More than 20% nulls - Consider advanced imputation or dropping
                #    # Remove rows with nulls - an option, check first the impact
                #     df = df.dropna(subset=[col])


            elif pd.api.types.is_categorical_dtype(df[col]) or pd.api.types.is_object_dtype(df[col]):
                if (df[col].isnull().sum() / len(df)) < 0.05:
                    df[col] = df[col].fillna(df[col].mode()[0]) # Impute with mode
                else:
                    df[col] = df[col].fillna("Missing")  # Replace with string "Missing"

    return df

# Assuming 'df' is your DataFrame
df = impute_nulls(df)
df.isnull().sum()
```

```python
# What is the distribution of departure delays across all flights?

# Histogram of Departure Delays
plt.figure(figsize=(10, 6))
sns.histplot(df['DEPARTURE_DELAY'], kde=True, bins=30)
plt.title('Distribution of Departure Delays')
plt.xlabel('Departure Delay (Minutes)')
plt.ylabel('Frequency')
plt.show()
```
