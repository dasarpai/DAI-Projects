```python
import csv
import subprocess




# Read employee data from the CSV file
with open('employee.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # Define the LaTeX content with placeholders
        name = row["Name"]
        print(name)

        latex_template = f"""
\\documentclass{{article}}
\\begin{{document}}

Dear \\textbf{ {name} },

Wishing you a Happy New Year filled with joy and success.

Sincerely,
Advaitananda
\\end{{document}}"""

        print(latex_template)

        # Save the LaTeX content to a .tex file
        with open(name+'new_year_wish.tex', 'w') as texfile:
            texfile.write(latex_template)

        # Compile the LaTeX document to a PDF
        subprocess.call(['pdflatex', name+'new_year_wish.tex'])

        # You can then email or distribute the generated PDF to the recipient
```
